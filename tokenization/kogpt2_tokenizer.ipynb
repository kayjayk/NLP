```python
with open('../Data/brunch/sentence_pickles/dict_sentences_210621-removed-1.pickle', 'rb') as f:
    tmp = pickle.load(f)
```

```python
all_sentences = []
```

```python
all_sentences = []
for topic in tmp.keys():
    for user_id in tmp[topic].keys():
        for sentence in tmp[topic][user_id]:
            all_sentences.append(sentence)
```

```python
all_sentences_to_id = []
```

```python
for sentence in tqdm.tqdm(all_sentences):
    a = tokenizer.encode(sentence)
    all_sentences_to_id.append(a)
```

```python
len(all_sentences_to_id)
```

* tf-idf 구현

```python
N = len(all_sentences) # 총 문장의 수

def tf(t, d):
    return d.count(t)

def idf(t):
    df = 0
    for sentence_to_id in all_sentences_to_id:
        df += t in sentence_to_id
    return log(N/(df + 1))

def tfidf(t, d):
    return tf(t,d)* idf(t)
```

```python
print(N)
```

```python
vocab = list(range(50000))
```

* idf 만 생성

```python
result = []
for j in tqdm.tqdm(range(len(vocab))):
    t = vocab[j]
    result.append(idf(t))
```

```python
idf_ = pd.DataFrame(result, index = vocab, columns = ["IDF"])
idf_
```

```python
idf_['idx'] = range(50000)
```

```python
idf_.to_csv('./kogpt2_tokenizer_idf.csv',index=None,)
```

```python
sorted_idf = idf_.sort_values(by=['IDF'])
```

```python
for idx, row in sorted_idf.iterrows():
    a = tokenizer.decode([idx])
    print(row['IDF'], a)
```

* tf matrix 생성 (100분 정도 소요)

```python
result = []
for i in tqdm.tqdm(range(N)): # 각 문장에 대해서 아래 명령을 수행
    result.append([])
    sentence = all_sentences_to_id[i]
    for j in range(len(vocab)):
        t = vocab[j]        
        result[-1].append(tf(t, sentence))
```

```python
tf_ = pd.DataFrame(result, columns = vocab)
tf_
```
